# Analysis of Wikidata Triplet Extraction (Beta)

```python
import pandas as pd
import seaborn as sns
```

```python
wikidata_triplets_df = pd.read_csv("../data/wikidata_triplets_beta.csv")
wikidata_triplets_negatives_df = pd.read_csv(
    "../data/wikidata_triplets_negative_beta.csv"
)
```

```python
wikidata_triplets_df.head()
```

```python
wikidata_triplets_negatives_df.head()
```

## Positive Triplet Extraction

```python
# Number of unique p_names in the dataset
print(
    f"Number of unique p_names in the dataset: {wikidata_triplets_df.p_name.nunique()}"
)
```

```python
# Plot p_names in wikidata_triplets_no_tacred in a bar plot, ordered by frequency, for the 50 most frequent p_names
p_name_counts = wikidata_triplets_df.p_name.value_counts()
p_name_counts = p_name_counts[:30]

ax = sns.countplot(y="p_name", data=wikidata_triplets_df, order=p_name_counts.index)

ax.set_title("Top 30 p_names in wikidata_triplets_beta")
ax.set_xlabel("Frequency")
ax.set_ylabel("p_name")
```

```python
# Line plot with the number of triplets per p_name
p_name_counts = wikidata_triplets_df.p_name.value_counts()

ax = sns.lineplot(x=p_name_counts.index, y=p_name_counts.values)

ax.set_title("Number of triplets per p_name in wikidata_triplets_beta")
ax.set_xlabel("p_name")
ax.set_ylabel("Number of triplets")

# Dont show the names of the p_names in the x-axis legend
ax.set_xticklabels([])
ax.set_xticks([])
```

```python
# Countplot of p_value_type
ax = sns.countplot(x="p_value_type", data=wikidata_triplets_df)

ax.set_title("p_value_type in wikidata_triplets_beta")
ax.set_xlabel("Frequency")
ax.set_ylabel("p_value_type")
```
