# Wikidata Triplet2Text Dataset Exploration

```python
import pandas as pd
```

## Load the dataset

```python
wikidata_triplets_df = pd.read_csv("../data/wikidata_triplets.csv")
wikidata_triplets_df.head()
```

## Generating positive samples

```python
for i, row in wikidata_triplets_df.iterrows():
    print(f"Input: {row['q_name']} {row['p_name']} {row['p_value']}")
    prompt = f"You are a AI assistant aimed to generate a sentences given a triplet with a subject, a relation and the object of a relation. Please generate a sentence that describes the triplet. If there is a date and time, please ignore the time. Here is a triplet: {row['q_name']} {row['p_name']} {row['p_value']}."
    print(prompt)
    print()
```
